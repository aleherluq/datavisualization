```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear DataFrame de ejemplo
datos = pd.DataFrame({'mes': [1, 2, 3, 1, 2, 3],
                      'año': [2022, 2022, 2022, 2023, 2023, 2023],
                      'valor': [10, 20, 30, 40, 50, 60]})

# Mapear números de mes a nombres de mes
nombres_meses = datos['mes'].map({1: 'Ene', 2: 'Feb', 3: 'Mar'})

# Reemplazar columna 'mes' con nombres de mes
datos['mes'] = nombres_meses

# Agrupar por mes y año y calcular la suma de los valores
datos_agrupados = datos.groupby(['mes', 'año']).sum()

# Restablecer los índices
datos_agrupados = datos_agrupados.reset_index()

# Crear gráfica de barras por año
plt.figure(figsize=(10, 6))
ancho_barras = 0.2
colores = ['b', 'g', 'r']  # Colores para cada año
años = datos_agrupados['año'].unique()

for i, año in enumerate(años):
    datos_año = datos_agrupados[datos_agrupados['año'] == año]
    posiciones_barras = range(i, len(datos_agrupados), len(años))
    plt.bar(posiciones_barras, datos_año['valor'], width=ancho_barras,
            label=str(año), color=colores[i % len(colores)])

plt.xlabel('Mes')
plt.ylabel('Suma de Valores')
plt.title('Suma de Valores por Mes y Año')
plt.legend(title='Año')
plt.xticks(range(len(datos_agrupados)), datos_agrupados['mes'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear DataFrame de ejemplo
datos = pd.DataFrame({'mes': [1, 2, 3, 1, 2, 3],
                      'año': [2022, 2022, 2022, 2023, 2023, 2023],
                      'valor': [10, 20, 30, 40, 50, 60]})

# Mapear números de mes a nombres de mes
nombres_meses = datos['mes'].map({1: 'Ene', 2: 'Feb', 3: 'Mar'})

# Reemplazar columna 'mes' con nombres de mes
datos['mes'] = nombres_meses

# Agrupar por mes y año y calcular la suma de los valores
datos_agrupados = datos.groupby(['mes', 'año']).sum()

# Restablecer los índices
datos_agrupados = datos_agrupados.reset_index()

# Crear gráfica de barras por mes y año
plt.figure(figsize=(10, 6))
ancho_barras = 0.2
colores = ['b', 'g', 'r']  # Colores para cada año
años = datos_agrupados['año'].unique()

for i, año in enumerate(años):
    datos_año = datos_agrupados[datos_agrupados['año'] == año]
    posiciones_barras = range(i, len(datos_agrupados), len(años))
    if i == 0:
        zorder = len(años)  # Año menor delante
    else:
        zorder = 0  # Año mayor detrás
    plt.bar(posiciones_barras, datos_año['valor'], width=ancho_barras,
            label=str(año), color=colores[i % len(colores)], zorder=zorder)

plt.xlabel('Mes')
plt.ylabel('Suma de Valores')
plt.title('Suma de Valores por Mes y Año')
plt.legend(title='Año')
plt.xticks(range(len(datos_agrupados)), datos_agrupados['mes'])
plt.show()
```

```python
df = pd.DataFrame({'mes': ['enero', 'enero','febrero', 'marzo', 'abril', 'mayo', 'junio'],
                   'valor': [10, 50, 20, 30, 40, 50, 60]})

# Crear un diccionario para mapear nombres de meses a números de mes
meses_numeros = {'enero': 1, 'febrero': 2, 'marzo': 3, 'abril': 4, 'mayo': 5, 'junio': 6}

# Mapear nombres de meses a números de mes
df['mes_num'] = df['mes'].map(meses_numeros)
print(df)
# Agrupar por la columna 'mes_num' y sumar los valores en cada grupo
df_grouped = df.groupby(['mes_num','mes'])['valor'].sum()
print(df_grouped)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear DataFrame de ejemplo
datos = pd.DataFrame({'mes': [1, 2, 3, 1, 2, 3],
                      'año': [2022, 2022, 2022, 2023, 2023, 2023],
                      'valor': [10, 20, 30, 40, 50, 60]})

# Agrupar por mes y año y calcular la suma de los valores
datos_agrupados = datos.groupby(['mes', 'año']).sum()

# Restablecer los índices
datos_agrupados = datos_agrupados.reset_index()

# Crear gráfica de líneas por año
plt.figure(figsize=(10, 6))
for año in datos_agrupados['año'].unique():
    datos_año = datos_agrupados[datos_agrupados['año'] == año]
    plt.plot(datos_año['mes'], datos_año['valor'], label=año)
plt.xlabel('Mes')
plt.ylabel('Suma de Valores')
plt.title('Suma de Valores por Mes y Año')
plt.legend(title='Año')
plt.show()
```
